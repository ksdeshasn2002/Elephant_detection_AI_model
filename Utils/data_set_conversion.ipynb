```python
import os
import shutil
import xml.etree.ElementTree as ET

# Paths
train_dir = "train"   # your folder containing images + xml
output_dir = "dataset_binary"

# Create output folders
os.makedirs(os.path.join(output_dir, "elephant"), exist_ok=True)
os.makedirs(os.path.join(output_dir, "non_elephant"), exist_ok=True)

# Loop over XML files
for file in os.listdir(train_dir):
    if not file.endswith(".xml"):
        continue
    
    xml_path = os.path.join(train_dir, file)
    tree = ET.parse(xml_path)
    root = tree.getroot()

    # Image filename (same as XML but .jpg)
    filename = root.find("filename").text
    img_path = os.path.join(train_dir, filename)

    # Get all labels in this image
    labels = [obj.find("name").text.lower() for obj in root.findall("object")]

    # Decide class
    if "elephant" in labels:
        target_folder = "elephant"
    else:
        target_folder = "non_elephant"

    # Copy image into new folder
    if os.path.exists(img_path):  # safety check
        shutil.copy(img_path, os.path.join(output_dir, target_folder, filename))

print("✅ Dataset conversion done!")
```
